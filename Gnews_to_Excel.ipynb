```python
import calendar
import json
import time
import urllib.request

apikey = "#"
q = "AAPL"
start_date = "2018-01-01"
end_date = "2019-06-03"
polarity = []
published_date = []
all_data = []

start_year, start_month, start_day = map(int, start_date.split("-"))
end_year, end_month, end_day = map(int, end_date.split("-"))

for year in range(start_year, end_year + 1):
    start_month_range = 1 if year == start_year else start_month
    end_month_range = 12 if year == end_year else end_month

    for month in range(start_month_range, end_month_range + 1):
        dias_no_mes = calendar.monthrange(year, month)[1]
        start_day_range = 1 if year == start_year and month == start_month else start_day
        end_day_range = dias_no_mes if year == end_year and month == end_month else dias_no_mes + 1

        for day in range(start_day_range, end_day_range):
            day_mean = []
            time.sleep(1)
            formatted_date = f"{year}-{month:02}-{day:02}"
            url = f"https://gnews.io/api/v4/search?q={q}&lang=en&country=us&max=25&apikey={apikey}&from={formatted_date}T00:00:01Z&to={formatted_date}T23:59:00Z"

            with urllib.request.urlopen(url) as response:
                data = json.loads(response.read().decode("utf-8"))
                articles = data["articles"]
                all_data.extend(articles)

                for article in articles:
                    print(f"Title: {article['title']}")
                    print(f"Source: {article['source']['name']}")
                    print(f"Description: {article['description']}")
                    print(f"PublishedAt: {article['publishedAt']}")
                    print("\n")
```

```python
import pandas as pd
a = pd.DataFrame(all_data)
```

```python
def export_dataframe_to_excel(dataframe, variable_name, kind_info):
    # Criar o caminho da pasta
    folder_path = '/content/drive/MyDrive/Estudos/Pesquisa - Diversificação de Ações/Long - Short/Algoritmos/News_Code/' + variable_name

    # Criar a pasta
    os.makedirs(folder_path, exist_ok=True)

    # Definir o caminho do arquivo XLSX
    file_path = os.path.join(folder_path, f'{variable_name}_{kind_info}.xlsx')

    # Exportar o DataFrame para XLSX
    dataframe.to_excel(file_path, index=True)

    print('DataFrame exportado com sucesso para:', file_path)
```
