```python
import requests
import json
import urllib.request
```

```python
apikey = "#"
q = "Bradesco"
url = f"https://gnews.io/api/v4/search?q={q}&lang=pt&country=br&max=10&apikey={apikey}&from=2023-03-20T12:30:00Z&to=2023-03-21T23:59:00Z"

with urllib.request.urlopen(url) as response:
    data = json.loads(response.read().decode("utf-8"))
    articles = data["articles"]


    for a in range(len(articles)):
        print(f"Title: {articles[a]['title']}")
        print(f"source: {articles[a]['source']['name']}")
        print(f"Description: {articles[a]['description']}")
        print(f"publishedAt: {articles[a]['publishedAt']}")
        print("\n")



```

```python
###Sentic Net Analysis
```

```python
polarity = []

for a in range(len(articles)):
    input_text = f"{articles[a]['description']} {articles[a]['title']}"
    url = f"http://sentic.net/api/pt/#.py?text={input_text}"
    response = requests.get(url)
    # Parse the JSON response and extract the polarity value
    polarity_value = json.loads(response.content.decode())
    # Append the polarity value to the polarity array
    polarity.append(polarity_value)
```
